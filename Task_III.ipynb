```python
df3 =pd.DataFrame(df1.groupby('Id')['total_active_mins'].sum())
df3
```

```python
cursor.execute("select * from (select *,(VeryActiveMinutes+FairlyActiveMinutes+LightlyActiveMinutes) as total_active_minutes from ineuron.fitbitdata order by 
(VeryActiveMinutes+FairlyActiveMinutes+LightlyActiveMinutes) desc limit 3) A order by A.total_active_minutes asc limit 1")
```

### 11. who is the 5th most laziest person avilable in dataset find it out

```python
lazydf=pd.DataFrame(df1.groupby('Id')['SedentaryMinutes'].sum())
```

```python
lazydf[lazydf['SedentaryMinutes'].rank( method='average',ascending = False)==5.0]
```
